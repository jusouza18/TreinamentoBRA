### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
import numpy as np
arr = np.random.randint(1, 51, size=10)

media = np.mean(arr)
desvio_padrao = np.std(arr)

print(arr)
print(media)
print(desvio_padrao)
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
import numpy as np

arr = np.random.rand(10)
arr[arr <= 0.5] = np.sqrt(arr[arr <= 0.5])
print(arr)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
import numpy as np

matriz = np.random.rand(3, 3, 3)

valor_max = np.amax(matriz, axis = 1)

print('Matriz:\n',matriz)
print('Valor Máximo:\n',valor_max)
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
import numpy as np

matriz = np.random.rand(5, 5)

media = np.mean(matriz, axis=1)

matriz_linha_media = matriz - media.reshape((-1,1))

print("Matriz:\n", matriz)
print("Média Linhas:\n", media)
print("Matriz Subtração:\n", matriz_linha_media)
```
